```python
def draw_me(f):
    all_surv = draw_survival_curves(mut.features.ix[f], surv, ann='p', show=True, labels=['WT', 'Mutated'])
    try:
        split_by_stage = draw_survival_curves(mut.features.ix[f], surv, stage_1, ann='p', 
                                              labels=['WT', 'Mutated'], show=True, filename='tmp2.png')
    except:
        return all_surv
    try:
        figsize=(6,4)
        pathway_plot(mut.df.ix[run.gene_sets[f]], plt.gca())
        plt.tight_layout()
        plt.savefig('tmp1.png', dpi=75, bbox_inches=0, pad_inches=0)
        plt.close('all')
        return stack([side_by_side(['tmp.png', 'tmp1.png']), split_by_stage])
    except:
        return stack([all_surv, split_by_stage])
    
s = stack([draw_me(f) for f in mut_res.index[:10]])
s
```
