#HNSC Copy Number Event Screen 
* This is a screen for survival associated copy-number event features. 
* I am using the results of the [GISTIC2](http://www.broadinstitute.org/cgi-bin/cancer/publications/pub_paper.cgi?mode=view&paper_id=216&p=t) analysis for the event calls. 
* Briefly their encoding for CN events is as follows:
    * -2 homozygous loss
    * -1 heterozygous loss
    * 0 diploid
    * 1 amplification
    * 2 high grade amplification
* There are two types of features: 
    * Lesions are events that occur on a broad level  
    * Gene level events are localized to a gene
* Since gene level events on a shared copy-number band are usually highly correlated I combine them for the sake of limiting the number of tests preformed.

##Screen for survival associated features  
* First pass is a log-rank test on all features
* Next we run the full model with covariates  
* We use a combination of the two tests to filter down the candidate associations and pull out the top biomarkers
* Here the log-rank test is really underpowered.  
* The deletion-rate covariate helps a lot to explain away many of the patients who have very high mutation rates and good prognosis.
* The age covariate often explains patients who are in a good-prognosis group but have bad outcomes.
* The results of the full model are kind of hard to interperet and non-necessarily relavant as we get a lot of features that help to explain rate effect, ect. and thus improve the model but we don't exactly know what to do with these things for now.   

```python
survival_test = 'survival_5y'
covariates = ['ageatinitialpathologicdiagnosis']
cov_df = global_vars.join(clinical.clinical, how='outer')
cov_df = cov_df[covariates]
cin = global_vars[('cna', 'chrom_instability')]
cov_df[('CIN')] = cin
surv = clinical.survival[survival_test]
test = SurvivalTest(surv, cov_df)
test.name = survival_test
test.check_featues = lambda s: s.sum() > 10
```

```python
df = (cn.features.ix['Deletion',keepers] < 0).astype(int)
df = df[df.sum(1) > 10]

lr = df.apply(log_rank, args=(surv,), axis=1)
lr = lr.sort('p')
lr_del = lr
```

```python
df = (cn.features.ix['Amplification',keepers] > 0).astype(str)
df = df[df.sum(1) > 10]

lr = df.apply(log_rank, args=(surv,), axis=1)
lr = lr.sort('p')
lr_amp = lr
```

```python
lr = pd.concat([lr_del, lr_amp], keys=['Deletion','Amplification'])
lr['q'] = bhCorrection(lr.p)
```

```python
lr.sort('p').head()
```

###Now for the full test

```python
rate = global_vars[('cna', 'lesion_del')]
rate.name = 'rate'
cov_df = pd.concat([age, rate], axis=1).dropna()
test = SurvivalTest(surv, cov_df)

df = (cn.features.ix['Deletion',keepers] < 0).astype(int)
df = df[df.sum(1) > 10]
cn_del_full = df.apply(test.full_test, 1)
```

```python
rate = global_vars[('cna', 'lesion_amp')]
rate.name = 'rate'
cov_df = pd.concat([age, rate], axis=1).dropna()
test = SurvivalTest(surv, cov_df)

df = (cn.features.ix['Amplification',keepers] > 0).astype(int)
df = df[df.sum(1) > 10]
cn_amp_full = df.apply(test.full_test, 1)
```

```python
cn_full = pd.concat([cn_del_full, cn_amp_full], keys=['Deletion','Amplification'])
cn_full['LR_q'] = bhCorrection(cn_full.LR)
```

* p16 seems to get bumped up there seems to be some differential effect depending on the deletion rate.  This is likely due in part to its interaction with TP53. 
* 3p seems to preform much better than p16 in the simple model, I think p16 may be able to explain some of the p53 effect while 3p is too heavily redundant with p53?

####Here we build up a matrix of descriptive statistics for the genes in our copy number arm
* exp_cn: change in expression across diploid and deleted patients (Kruskal-Wallis test)
* exp_tumor: change in expression across normal and matched tumor tissue
    * Kruskal-Wallis test
    * Note uneven sample sizes due to unavailibility of matched expression samples for all patients
    * Using only matched samples with a paired test, yeild similar results but we are just using all the data and treating them as poplulations for now 
* exp_p53: Spearman correlation of expression with TP53 expression
    * might be cheating a little because it has been published that FHIT is coexpressed with TP53 ( http://www.ncbi.nlm.nih.gov/pubmed/15313915 )
* exp_survp: Univariate Cox test for association of experession with survival
    * The idea here is that the expression change driven by the copy number loss should influence survival

```python
loc['exp_cn'] = pd.Series({f: kruskal_pandas(vec < -.1, tumor.ix[f[0]].replace(-3, nan))['p'] 
                           for f,vec in submat.iterrows() 
                           if f[0] in tumor.index}).order()
loc['exp_cn'] = -np.log10(loc['exp_cn'])

loc['exp_tumor'] = pd.Series({f: kruskal_pandas(normals, rnaSeq.ix[f[0]])['p'] 
                           for f, vec in submat.iterrows() 
                           if f[0] in tumor.index}).order()
loc['exp_tumor'] = -np.log10(loc['exp_tumor'])

loc['exp_p53'] = pd.Series({f: spearman_pandas(tumor.ix[f[0]], tumor.ix['TP53'])['p'] 
                           for f,vec in submat.iterrows() 
                           if f[0] in tumor.index}).order()
loc['exp_p53'] = -np.log10(loc['exp_p53'])
e = tumor.ix[submat.index.get_level_values(0)]
e.index = submat.index
e = e.dropna()
loc['exp_survp'] = pd.DataFrame({g: test.first_pass((s-s.mean()/s.std())) for g,s in e.iterrows()}).T['p']
loc['exp_survp'] = -np.log10(loc['exp_survp'])

e = meth.df.ix[submat.index.get_level_values(0)]
e.index = submat.index
e = e.dropna()
loc['meth_survp'] = pd.DataFrame({g: test.first_pass((s-s.mean()/s.std())) for g,s in e.iterrows()}).T['p']
loc['meth_survp'] = -np.log10(loc['meth_survp'])
```
