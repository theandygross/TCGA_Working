```python
def series_scatter(s1, s2, ax=None, filename=None, **plot_args):
    if ax is None:
        fig, ax = plt.subplots(1,1, figsize=(6,4))
    ax.scatter(*match_series(s1, s2), alpha=.5, s=75, **plot_args)
    ax.set_xlabel(s1.name)
    ax.set_ylabel(s2.name)
    if filename is not None:
        fig.savefig(filename)
```

```python
ts = (ts - ts.mean()) / ts.std()
ercc1 = (ercc1 - ercc1.mean()) / ercc1.std()
```

```python
def quadrant_plot(g1, g2, thresh=.25):
    fig, ax = subplots(1,1)
    g1 = (g1 - g1.mean()) / g1.std()
    g2 = (g2 - g2.mean()) / g2.std()
    series_scatter(g1[g1 > thresh], g2[g2 > thresh], ax=ax, color='b')
    series_scatter(g1[g1 < -thresh], g2[g2 > thresh], ax=ax, color='r')
    series_scatter(g1[g1 < -thresh], g2[g2 < -thresh], ax=ax, color='g')
    series_scatter(g1[g1 > thresh], g2[g2 < -thresh], ax=ax, color='y')
```

```python
quadrant_plot(ts, ercc1, thresh=.25)
```

```python
def quadrant_strat(g1, g2, thresh=.25):
    g1 = (g1 - g1.mean()) / g1.std()
    g2 = (g2 - g2.mean()) / g2.std()
    
    f = (((g1 > thresh) * (g2 > thresh) * 1) +
     ((g1 < -thresh) * (g2 > thresh) * 2) +
     ((g1 < -thresh) * (g2 < -thresh) * 3) +
     ((g1 > thresh) * (g2 < -thresh) * 4))
    f.name = 'quadrants'
    return f
```
