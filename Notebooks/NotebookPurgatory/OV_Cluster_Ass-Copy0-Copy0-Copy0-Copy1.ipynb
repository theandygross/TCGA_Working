```python
from rpy2.robjects.packages import importr
```

```python
samr = importr('samr')
```

```python
rppa = cancer.load_data('RPPA')
mut = cancer.load_data('MAF')
mut.uncompress()
rnaSeq = cancer.load_data('mRNASeq')
#meth = cancer.load_data('Methylation')
cn = cancer.load_data('CN_broad')
cn.uncompress()
arm = pd.read_table(run.data_path + '/analyses/{}/CopyNumber_Gistic2/broad_values_by_arm.txt'.format(cancer.name), index_col=0)
arm = arm.rename(columns=lambda s: s[:12])
```

```python
from Data.Firehose import *
```

```python
#x = pd.concat([rppa.features, mut.features, rnaSeq.features, meth.features, cn.features], keys=['rppa','mut','rna','meth','cn'])
x = pd.concat([rppa.features, mut.features, rnaSeq.features,cn.features, arm], keys=['rppa','mut','rna','cn', 'cn_arm'])
#x = arm
features = x.index
x.index = map(str, x.index)
x = x.ix[:,s.index].dropna(1, thresh=x.shape[1] - 80).astype(float)
#x = x.ix[:,s.index].dropna(1, how='all').astype(float)
dim = x.shape
pat = x.columns
x = com.convert_to_r_matrix(x)
```

```python
results = {}
for i in range(1,5):
    y = (s == i) + 1
    y = robjects.r.c(*list(y.ix[pat].dropna()))
    res = samr.SAM(x=x, y=y, resp_type='Two class unpaired', test='wilcoxon', nperm=1000)
    r = []
    r_n = []
    try:
        m = res.rx2('siggenes.table').rx2('genes.lo')
        lo = com.convert_robj(robjects.r('as.data.frame')(m))
        lo['Gene Name'] = lo['Gene Name'].map(lambda idx: features[int(idx)-1])
        r_n += ['lo']
        r += [lo]
    except:
        print i
    try:   
        m = res.rx2('siggenes.table').rx2('genes.up')
        up = com.convert_robj(robjects.r('as.data.frame')(m))
        up['Gene Name'] = up['Gene Name'].map(lambda idx: features[int(idx)-1])
        r_n += ['up']
        r += [up]
    except:
        print i
    results[i] = pd.concat(r, keys=r_n)
    
```

```python
r = pd.concat(results.values(), keys=results.keys())
r['score_abs'] = -r['Score(d)'].astype(float).abs()
r['q-value(%)'] = r['q-value(%)'].astype(float)
r['Fold Change'] = r['Fold Change'].astype(float)
r1 = r[r['Gene Name'].map(lambda s: s[0] != 'mut')]
```
