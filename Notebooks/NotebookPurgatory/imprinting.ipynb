```python
from pandas import read_table
import pandas as pd
from matplotlib_venn import venn3
from rpy2 import robjects

pd.set_option('precision', 3)

def process(df):
    df.name = df.name.str.strip()
    df = df.set_index('name')
    df = df.groupby(level=0).first()
    return df

from Processing.Helpers import bhCorrection
from scipy.stats import hypergeom
def hyper_test(vec, gene_set, q=.1, return_vals=False):
    M = len(vec)
    n = len(gene_set)
    significant = set(vec[vec < q].index)
    N = len(significant)
    x = len(significant.intersection(set(gene_set)))
    if not return_vals:
        return hypergeom.sf(x, M, n, N)
    else:
        r = {'p': hypergeom.sf(x, M, n, N), 'x': x, 'M': M, 'n':n, 'N': N}
        return pd.Series(r)
```

```python
arc = pd.read_table(path + 'ARC_GeneSummedReads_additivemodel_results_decode copy.txt', index_col=0)
arc = process(arc)
dr = pd.read_table(path + 'DR_CCHCHH_additivemodel_results_real_GeneSummed_decoded copy.txt', index_col=0)
dr = process(dr)
liver = pd.read_table(path + 'LIVER_SummedReadsGene_AdditiveModel_results_fixed_decoded copy.txt', index_col=0)
liver = process(liver)
df = pd.concat([arc, dr, liver], keys=['ARC','DR','LIVER'], axis=1)
```
