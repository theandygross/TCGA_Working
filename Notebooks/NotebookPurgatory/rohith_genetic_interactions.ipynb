```python
def combine(a,b, fill=False):
    '''
    Combine two categorical features.
    '''
    if not fill:
        combo = (a*1.).add(b*2.)
    else:
        combo = (a*1.).add(b*2., fill=0)
    combo = combo.dropna()
    if not a.name:
        a.name = 'first'
    if not b.name:
        b.name = 'second'
    if a.name != b.name:
        combo = combo.map({0: 'neither', 1: a.name, 2: b.name, 3:'both'})
    else:
        combo = combo.map({0: 'neither', 1: 'first', 2: 'second', 3:'both'})
    return combo
```

```python
def overlap(a,b):
    try:
        aa = mut_genes.ix[a].T.stack() > 0
        bb = mut_genes.ix[b].T.stack() > 0
        idx = aa.index.union(bb.index)
        c = combine(aa.ix[idx]==True, bb.ix[idx]==True)
        c = c.groupby(level=0).value_counts().unstack().fillna(0)
        return c
    except:
        return
```
