```python
from Stats.Regression import *
def stratified_regression(target, feature, covariate, print_model=False):
    target = target.astype(int)
    target, feature, strata = map(sanitize_for_r, [target, feature, covariate])
    fmla = '{} ~ {} + {}'.format(target.name, feature.name, covariate.name)
    fmla = robjects.Formula(fmla)
    df_r = process_factors([covariate, target, feature], False)
    fit = glm(fmla, data=df_r, family='binomial')
    
    fmla = '{} ~ {}'.format(target.name, covariate.name)
    fmla = robjects.Formula(fmla)
    fit_null = glm(fmla, data=df_r, family='binomial')
        
    f_stat = robjects.r.anova(fit, fit_null, test='Chisq')[3][1]
    p = robjects.r.anova(fit, fit_null, test='Chisq')[4][1]
    
    return pd.Series({'F': f_stat, 'p': p})

def process_factors(cov, norm=True):
    '''
    Coerce covariates and feature into format suitable for R's
    survival functions. 
    '''
    if type(cov) == pd.Series:
        cov = pd.concat([cov], axis=1)
    elif type(cov) == list:
        assert map(type, cov) == ([pd.Series] * len(cov))
        cov = pd.concat(cov, axis=1)
        
    
    c_real = cov.ix[:,cov.dtypes.isin([np.dtype(float), np.dtype(int)])]
    if norm:
        c_real = (c_real - c_real.mean()) / c_real.std()
    cov[c_real.columns] = c_real
    df = cov
    df = df.groupby(level=0).first()
    df = df.dropna()
    df = convert_to_r_dataframe(df)
    return df
```
