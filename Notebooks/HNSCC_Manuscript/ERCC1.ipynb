```python
from Data.ProcessClinicalDataPortal import *
def format_survival_data(timeline, clin):
    timeline = timeline.applymap(try_float)
    timeline['days'] = timeline.astype(float).max(1)
    
    # deceased = clin.vitalstatus.dropna().isin(['DECEASED','Dead','deceased', False])
    deceased = timeline.daystodeath.isnull() == False
    
    # timeline = events.combine_first(timeline).astype(float)
    timeline[timeline < -1] = np.nan
    survival = pd.concat([timeline.days, deceased], keys=['days', 'event'],
                         axis=1)
    survival = survival.dropna().stack().astype(float)
    
    f_vars = ['days', 'daystonewtumoreventafterinitialtreatment',
              'daystotumorprogression', 'daystotumorrecurrence',
              'daystonewtumoreventafterinitialtreatment',
              'daystonewtumoreventafterinitialtreatment']
    followup_cols = list(timeline.columns.intersection(f_vars))
    pfs = timeline[followup_cols].min(1)
    # pfs = pd.concat([timeline.days, timeline[followup_cols]], axis=1).min(1)
    event = ((pfs < timeline.days) + deceased) > 0
    pfs = pd.concat([pfs, event], keys=['days', 'event'], axis=1)
    pfs = pfs.dropna().stack().astype(float)
    
    survival = pd.concat([survival, pfs],
                         keys=['survival', 'event_free_survival'],
                         axis=1)
    return timeline, survival

last_drug = clinical.drugs.daystodrugtherapyend.dropna().groupby(level=0).max()
last_drug.name = 'days_to_last_drug'
timeline = clinical.timeline.join(last_drug)

f = '{}/Followup/{}/clinical_radiation_{}.txt'.format(params['OUT_PATH'], cancer.name, cancer.name.lower())
t = pd.read_table(f, na_values=['[Not Available]','[Completed]'], index_col=0)
rads = t[['days_to_radiation_therapy_end','days_to_radiation_therapy_end']].max(1).dropna()
rads = rads.groupby(level=0).max()
rads.name = 'days_to_last_rads'

timeline = timeline.join(rads)

timeline['days'] = pd.concat([timeline.days, last_drug],1).max(1)
timeline, survival = format_survival_data(timeline, clinical.clinical)
clinical.survival = survival
clinical.artificially_censor(5)

surv = clinical.survival.survival_5y
```

```python
def quadrant_plot(g1, g2, thresh=.25):
    fig, ax = subplots(1,1)
    g1 = (g1 - g1.mean()) / g1.std()
    g2 = (g2 - g2.mean()) / g2.std()
    series_scatter(g1[g1 > thresh], g2[g2 > thresh], ax=ax, color='b')
    series_scatter(g1[g1 < -thresh], g2[g2 > thresh], ax=ax, color='r')
    series_scatter(g1[g1 < -thresh], g2[g2 < -thresh], ax=ax, color='g')
    series_scatter(g1[g1 > thresh], g2[g2 < -thresh], ax=ax, color='y')
    
def quadrant_strat(g1, g2, thresh=.25):
    g1 = (g1 - g1.mean()) / g1.std()
    g2 = (g2 - g2.mean()) / g2.std()
    
    f = (((g1 > thresh) * (g2 > thresh) * 1) +
     ((g1 < -thresh) * (g2 > thresh) * 2) +
     ((g1 < -thresh) * (g2 < -thresh) * 3) +
     ((g1 > thresh) * (g2 < -thresh) * 4))
    f.name = 'quadrants'
    return f
```
