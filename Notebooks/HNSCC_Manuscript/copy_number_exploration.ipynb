# Exploration of Chromosomal Instability

* We know chromosomal instability has a large role in cancer, lets see if we can find any good correlates 
* For chromosomal instability I'm using the total number of amplified or deleted lesions

```python
cin = ((cn.features.ix['Amplification'] > 0).sum() + 
       (cn.features.ix['Deletion'] > 0).sum())
cin.name = 'Chromosomal Instability'
```

```python
draw_survival_curves(to_quants(cin, std=1, labels=True), surv, hpv_inferred.dropna())
```

```python
violin_plot_pandas(hpv_inferred, cin)
```

```python
df = mut.df.ix[(mut.df.sum(1).order() > 5)]
tt = lambda a,b: kruskal_pandas(b,a)
s = screen_feature(cin, tt, df.clip_upper(1.))
s.head(10)
```

## Screening for copy-number correlates of chromosomal instability 

* These are all fragile sites usually around large genes ( http://www.sciencedirect.com/science/article/pii/S0304383505008232, http://www.sciencedirect.com/science/article/pii/S0168952512000303 )

```python
df = cn.features.ix['Deletion']
df.index = df.index.get_level_values(0)
screen_feature(cin, rev_kruskal, df < 0).head()
```

```python
df = cn.features.ix['Amplification']
df.index = df.index.get_level_values(0)
screen_feature(cin, tt, df > 0).head()
```

```python
fig, axs = subplots(1,2, figsize=(12,4))
violin_plot_pandas(cn.features.ix[('Deletion','5q11.2','Lesion')], cin, ax=axs[0])
violin_plot_pandas(cn.features.ix[('Deletion','3p14.2','Lesion')], cin, ax=axs[1])
```

####Is this just an artifact of chromosomal instablity?
* We screen the number of lesions deleted and amplified to see if thats what we are seeing 
* It seems like the 3p14 effect is stronger than an optimal threshold on chromosomal instability

```python
del_3p = cn.features.ix[('Deletion','3p14.2','Lesion')] < 0
```

```python
fig, axs = subplots(1,3, figsize=(12,4))
cin = cin.ix[keepers_o]
aa = pd.DataFrame({x: log_rank(cin >= x, surv) for x in arange(1, cin.max()-1)}).T
aa.dropna().chi2.plot(ax=axs[0])
axs[0].set_ylabel('Hazzard Ratio')
axs[0].set_xlabel('# of Lesions Deleted')
draw_survival_curve(((cin<6)*1.).map({True: '# Events < 6', False: '# LOH >= 6'}), surv, ax=axs[1])
draw_survival_curve(((del_3p==0)*1.).map({False: '3p14.2 +/-', True: ' 3p14.2 +/+'}), surv, ax=axs[2])
```

```python
cin = (cn.df.ix['Deletion'] < 0).sum()
cin = cin.ix[keepers_o]
cin.name = '# Deletions'

fig, axs = subplots(1,3, figsize=(12,4))
aa = pd.DataFrame({x: log_rank(cin < x, surv) for x in arange(1, cin.max()-1)}).T
aa.dropna().chi2.plot(ax=axs[0])
axs[0].set_ylabel('Hazzard Ratio')
axs[0].set_xlabel('# of Lesions Deleted')
draw_survival_curve(((cin<11)*1.).map({True: '# Events < 11', False: '# LOH >= 11'}), surv, ax=axs[1])
draw_survival_curve(((del_3p==0)*1.).map({False: '3p14.2 +/-', True: ' 3p14.2 +/+'}), surv, ax=axs[2])
```

# Extended Data Figure 3: Chromosome 3p Characterization

## Extended Data Figure 3a: HPV- Patients

```python
coords = pd.read_csv('../Extra_Data/gene_coords.csv', index_col=0)

cn_by_gene = FH.get_gistic_gene_matrix(run.data_path, cancer.name)
cn_by_gene = cn_by_gene.reset_index(level=[0,1])
cn_by_gene = cn_by_gene.join(coords).set_index(['chrom','start', 'Cytoband','Locus ID'], append=True)
cn_by_gene = cn_by_gene.ix[:,keepers_o].dropna(axis=1)

ch_3 = cn_by_gene.xs('3', level='chrom')
p_arm = pd.Series(ch_3.index.get_level_values('Cytoband').map(lambda s: 'p' in s), ch_3.index)
p_arm = ch_3[p_arm].sortlevel('start')
p_arm = p_arm.ix[:-1]
```

```python
#df = rna.df.xs('01', 1, 1).ix[p_arm.index.get_level_values(0)].dropna()#
#df = rna.features.ix['real'].ix[p_arm.index.get_level_values(0)].dropna()
df = rna.df.xs('01', 1, 1).ix[p_arm.index.get_level_values(0)].dropna()
#df = df.ix[screen_feature(del_3p, kruskal_pandas, df).p < .05]
df = df.ix[:,keepers_o].dropna(1)
test = lambda s: get_cox_ph_ms(surv, s, [age, old], interactions='just_first')
exp = df.apply(test, axis=1)
exp = exp.sort_index()
exp.index = p_arm.sort_index().ix[exp.index].index

exp2 = -exp.LR.map(log10)
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = plt.subplot2grid((7,1), (0,0), rowspan=3, colspan=1)
ax2 = plt.subplot2grid((7,1), (3,0), rowspan=3, colspan=1)
ax3 = plt.subplot2grid((7,1), (6,0), rowspan=1, colspan=1)

bands = p_arm.reset_index().groupby('Cytoband')
bands = bands.first().start.order()
starts = p_arm.index.get_level_values('start')
for i,(idx,v) in enumerate(bands.iteritems()):
    try:
        r = plt.Rectangle((v,0), bands.iloc[i+1]-v, 1, lw=2, 
                          alpha=.5, fc=colors[i%2])
    except:
        r = plt.Rectangle((v,0), starts[-1] - v, 1, lw=2, 
                          alpha=.5, fc=colors[i%2])
    ax3.add_artist(r)
ax3.set_xticks(bands)
b = [{0:s,1:''}[(i%2)] for i,s in enumerate(bands.index)]
ax3.set_xticklabels(b, rotation=270+45, ha='left')

ax1.scatter(starts, (p_arm < -.1).sum(1), s=40, c=colors[2], 
            edgecolor='grey', lw=.5, alpha=.5)
ax2.scatter(exp2.index.get_level_values('start'), exp2, s=40, 
            c=colors[4], edgecolor='grey', alpha=.7)

#ax1.set_ybound(215,255)
ax1.set_ylabel('# of Patients\nwith Deletion\n\n', ha='center', va='center')
ax2.set_ybound(-.5,4.)
ax2.set_ylabel('mRNA Survival\nAssociation\n(Log10 p)\n\n\n', ha='center', va='center')
ax3.set_yticks([])
fig.tight_layout()
for ax in [ax1, ax2, ax3]:
    ax.set_xbound(starts[0], 
                  starts[-1])
    ax.set_xticks(bands)

for n, ss in exp2.order().tail(3).iteritems():
    ax2.annotate(n[0], (n[1], ss), xytext=(15, -5), 
                 textcoords='offset points', 
                 arrowprops=dict(arrowstyle="->"))
ax1.set_xticklabels('')
ax2.set_xticklabels('')
ax2.set_yticks([0,1,2,3,4,5])
fig.tight_layout()
```
