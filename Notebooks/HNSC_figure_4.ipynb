```python
from matplotlib.patches import Circle
from matplotlib.patches import Rectangle

def plot_guy(ax, x, y, head='circle', **plot_args):
    an = array(linspace(0,2*pi,100))
    if head == 'circle':
        head = Circle((x, y+6), radius=2, color=plot_args['c'])
    else:
        head = Rectangle((x-2, y+6-2), 4, 4, color=plot_args['c'])
    body = array([0,0])+x, array([4,2])+y
    legs = array([-1,0,1])+x, array([0,2,0])+y
    arms = array([-1.5,0,1.5])+x, array([3.5,3,3.5])+y
    #eye1 = .2*cos(an)-.35 + x, .2*sin(an) + 5.2 + y
    #eye2 = .2*cos(an)+.35 + x, .2*sin(an) + 5.2 + y
    ax.plot(*(body + legs + arms), **plot_args)
    ax.add_patch(head)
```

```python
def plot_guys(guys, assignment, jitter=True, ax=None):
    if ax == None:
        fig, ax = subplots(figsize=(8,8))
    for i,(x, y) in enumerate(guys):
        if i >= len(assignment):
            continue
        if jitter:
            x += random_sample()*5
            y += random_sample()*5
        plot_args = dict(c=assignment[i][1], mec="b", mfc="w", lw=1, mew=3, ms=10)
        plot_guy(ax, x, y, head=assignment[i][0], **plot_args)
        ax.set_frame_on(False)
        ax.set_xticks([])
        ax.set_yticks([])
    ax.set_xbound(-10,200)
    ax.set_ybound(-10,200)
```

```python
assignments = pd.crosstab(to_quants(ff, std=1), two_hit).T.stack()
```

```python
heads = (['circle'] * assignments.sum(level=0).ix[0]) + (['square'] * assignments.sum(level=0).ix[1])
bodies = [[colors[int(idx[1] + 1)]] * count for idx,count in assignments.iteritems()]
bodies = list(flatten(bodies))
```

```python
assignment = zip(heads, bodies)
assignment = permutation(assignment)
```

```python
fig, axs = subplots(5,1, figsize=(8,40))

guys = [[x,y] for y in 170-arange(0,200,10) for x in arange(0,200,10)]
plot_guys(guys, assignment, ax=axs[0]);

guys = [[x,y] for y in 170-arange(0,200,10) for x in arange(0,150,10)]
assignment_1 = [a for a in assignment if a[0] == 'square']
plot_guys(guys, assignment_1, ax=axs[1]);

guys = [[x,y] for y in 170-arange(0,200,10) for x in arange(0,140,10)]
assignment_2 = [a for a in assignment if a[0] == 'circle']
plot_guys(guys, assignment_2, ax=axs[2]);

assignment_s = zip(heads, bodies)

guys = [[x,y] for y in 170-arange(0,200,10) for x in arange(0,90,10)]
assignment_2 = [a for a in assignment_s if a[0] == 'circle']
plot_guys(guys, assignment_2, jitter=False, ax=axs[3]);

guys = [[x,y] for y in 170-arange(0,200,10) for x in arange(0,100,10)]
assignment_1 = [a for a in assignment_s if a[0] == 'square']
plot_guys(guys, assignment_1, jitter=False, ax=axs[4]);
plt.savefig('/cellar/users/agross/Desktop/Figures/people.pdf', tranparent=True)
```
