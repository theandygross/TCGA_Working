####Build model with both features 
* We can see that the model does not explicitly include the interaction term.
* This is because of the constraint that a model with an interaction include both parent terms.
* If we force it, you can see the fit goes up but not enought to compensate for the added degree of freedom. 
* But in truth, we are saying that the function of these features are (mostly) redundant, in-so-much that the effect is felt only if BOTH are present.
* Reducing the model with this knowledge creates a simpler model while sacrificing very little of the fit (LR 33.7 to LR 31.9)
