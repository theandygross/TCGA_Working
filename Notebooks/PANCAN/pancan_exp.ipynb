```python
cancers = {}
clinical = {}
mut = {}
cn = {}
for cancer in run.cancers:
    if cancer == 'COAD':
        continue
    try:
        c = run.load_cancer(cancer)
        clinical[cancer] = c.load_clinical()
        mut[cancer] = c.load_data('Mutation')
        cn[cancer] = c.load_data('CN_broad')
        cancers[cancer] = c
    except:
        do_nothing = True
```

```python
def ttr(s):
    a,b = [l[1] for l in list(s.groupby(level=1)) if l[0] in ['01','11']]
    a.index = a.index.droplevel(1)
    b.index = b.index.droplevel(1)
    return ttest_rel(a,b)
```

```python
def ppp(s):
    s = s.unstack()[['01', '11']].dropna()
    return spearman_pandas(s['01'], s['11'])
```
