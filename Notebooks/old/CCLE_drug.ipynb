```python
fig, axs = subplots(6,4, figsize=(15,15))
drug = 'Paclitaxel'
for i,drug in enumerate(set(ss.index.get_level_values(1))):
    ax = axs[i%6][i/6]
    f = ss.xs(drug, level=1)
    for p,g in f.iterrows():
        x = eval(g['Doses (uM)'])
        y = eval(g['Activity Data (median)'])
        sd = eval(g['Activity SD'])
        ax.errorbar(log2(x), y, yerr=sd, ls='--', marker='s', label=p.split('_')[0])
    ax.set_xlabel('$Log_{2}$ Dose (uM)')
    ax.set_ylabel('Activity')
    ax.set_title('{} ({})'.format(drug, g['Target']))
axs[0][0].legend(numpoints=1)
fig.tight_layout()
fig.savefig('/cellar/users/agross/Desktop/drug_curves.pdf')
```
