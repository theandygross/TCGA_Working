```python
X = df2.T.as_matrix()
y = np.array(age.ix[df2.columns])
```

```python
alpha = 0.1
enet = ElasticNet(alpha=alpha, l1_ratio=0.7, normalize=True)

fit = enet.fit(X, y)
```

```python
model = pd.Series(fit.coef_, df2.index)
model = model[model != 0].order()
```

```python
pred_age = pd.Series(fit.predict(X), df2.columns, name='pred_age')
```
