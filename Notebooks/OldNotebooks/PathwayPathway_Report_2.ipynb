```python
def get_clinical_objects(cancer, data_types, test=False):
    clinical = {}
    for data_type in data_types:
        data_path = firehose_path + 'analyses__' + date_ + '/' + cancer + '/' + date + '/'
        clinical_obj = data_path + data_type + '_' + report_ext[data_type] + '/ClinicalObject.p'
        if os.path.isfile(clinical_obj):
            if test:
                clinical[data_type] = clinical_obj
            else:
                clinical[data_type] = pickle.load(open(clinical_obj, 'rb'))
    return clinical
```

```python
from pandas import DataFrame, Series
from Processing.Helpers import bhCorrection

def get_pathway_associations(mut_clustered, exp_clustered):
    p_vals =  DataFrame(dict((e,dict((m, anova(mut_vec, exp_vec)) 
                                     for m, mut_vec in mut_clustered.iterrows()))
                             for e, exp_vec in exp_clustered.iterrows()))
    m = p_vals.as_matrix()
    l = array(bhCorrection(Series(m.flatten())))
    q_vals = DataFrame(l.reshape(m.shape), index=p_vals.index, columns=p_vals.columns)
    return p_vals, q_vals

def within_clusters_p(mut_cluster, exp_cluster, mut, exp):
    exp_path = exp.assignments[exp.assignments==exp_cluster].index
    mut_path = mut.assignments[mut.assignments==mut_cluster].index
    p_vals =  Series(dict(((m,e), anova(mut_vec, exp_vec)) 
                           for m, mut_vec in mut.meta_matrix.iterrows()
                           for e, exp_vec in exp.pc.iterrows() 
                           if ((e in exp_path) and (m in mut_path))))
    return p_vals

def get_pathway_interaction_table(mut, exp):
    p_vals, q_vals = get_pathway_associations(mut.clustered, exp.clustered)
    p_vals = p_vals.dropna()
    
    def characterize(s):
        rows = {'Exp Group' : s.idxmin(),
                'Mut Group' : s.name,
                'p_val' : s.min(),
                'q_val' : q_vals.ix[s.name, s.idxmin()]}
        pathways = np.sort(within_clusters_p(s.name, s.idxmin(), mut, exp)).index[0]
        rows['Mut Pathway'] = pathways[0]
        rows['Exp Pathway'] = pathways[1] 
        return Series(rows)[['Mut Group', 'Exp Group', 'Mut Pathway', 'Exp Pathway', 
                             'p_val', 'q_val']]
    
    keep = q_vals.index[(q_vals < .05).sum(1) > 0]
    if len(keep) < 10:
        keep = q_vals.index[(q_vals < .01).sum(1) > 0]
        if len(keep) == 0:
            keep = np.sort(q_vals.min()).head(10).index
    int_table = p_vals.ix[keep].apply(characterize, axis=1).sort(columns='p_val')
    return int_table

def pathway_pathway_report(exp, mut, int_table, next_cancer, previous_cancer):
    path = exp.data_path + exp.data_type + '_' + mut.data_type
    if not os.path.isdir(path + '/' + FIG_EXT):
        os.makedirs(path + '/' + FIG_EXT)
    report = nz.newReport('Report for ' + exp.cancer)
    
    folder = exp.report_folder
    exp.report_folder = path
    report = generic_header(report, exp, next_cancer, previous_cancer)
    exp.report_folder = folder
    table = format_pathway_interaction_table(path, int_table, mut, exp)
    report = nz.addToResults(report, table)
    nz.writeReport(report, filename=path + '/index')
```

```python
from Processing.Helpers import bhCorrection
def generate_report(mut, exp, prev_cancer, next_cancer):
    exp_path = np.sort(exp.p_pathways.survival).index[0]
    exp_vec = exp.pc.ix[exp_path]
    mut_df = mut.meta_matrix
    
    p_vals =  pandas.Series(dict((m, anova(mut_vec, exp_vec)) for m, mut_vec in mut_df.iterrows()))
    q_vals = bhCorrection(np.sort(p_vals))
    
    path = exp.data_path + 'deletion_methylation'
    if not os.path.isdir(path + '/' + FIG_EXT):
            os.makedirs(path + '/' + FIG_EXT)
    report = nz.newReport('Report for ' + exp.cancer + ': ' + exp_vec.name.replace('_',' '))
    folder = exp.report_folder
    exp.report_folder = path
    report = generic_header(report, exp, prev_cancer, next_cancer)
    exp.report_folder = folder
    
    histo_file = exp.report_folder + '/clinical_figures/' + exp_vec.name + '.svg'
    histo_fig1 = nz.newFigure(histo_file, 'Eigen-patient for pathway.')
    curves_file = exp.report_folder + '/clinical_figures/' + exp_vec.name + '_survival.png'
    curves_fig2 = nz.newFigure(curves_file, 'Cox regression for pathway. (Hi and Low are +- 1 std).')
    report = nz.addToSummary(report, histo_fig1, curves_fig2)
    
    table_rows = dict(pathway=p_vals.index, p_val=p_vals, q_val=q_vals, age_p=mut.p_pathways.age)
    if 'survival' in mut.p_pathways:
        table_rows['survival_p'] = survival_p=mut.p_pathways.survival
    table = pandas.DataFrame(table_rows).sort('p_val')
    table = format_interaction_table(path, table, mut, exp_vec)
    report = nz.addToResults(report, table)
    nz.writeReport(report, filename=path + '/index')
```
