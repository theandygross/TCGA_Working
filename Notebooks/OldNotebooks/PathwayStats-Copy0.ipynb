```python


def mut_box(x, y, width=.7, height=.9, small=False):
    if small:
        height = height / 2
        y = y - height / 2
    width = width - .2
    height = height - .2
    rect = FancyBboxPatch([x - width / 2, y - height / 2], width, height,
                          facecolor='green', edgecolor='black', alpha=.8,
                          lw=2, mutation_scale=.33)
    return rect

def null_box(x, y, width=.7, height=.9):
    width = width - .2
    height = height - .2
    rect = FancyBboxPatch([x - width / 2, y - height / 2], width, height,
                     facecolor='grey', edgecolor='black', alpha=.5,
                     mutation_scale=.33)
    return rect

def memo_plot(df, ax=None):
    if ax == None:
        ax = plt.gca()
    ax.patch.set_facecolor('white')
    ax.set_aspect(3, 'box')
    ax.set_xticks([])
    ax.set_yticks(range(len(mat)))
    ax.set_yticklabels(df.index[::-1])
    
    W = df.as_matrix().T[:, ::-1]
    for (x,y),w in np.ndenumerate(W):
        if w == 0: 
            rect = null_box(x, y)
        else:     
            rect = null_box(x, y)
            ax.add_patch(rect)
            rect = mut_box(x, y) 
        ax.add_patch(rect)
    ax.set_xbound(-.5, len(W)-.5)
    ax.set_ybound(-.5, len(W[0])-.5)
    
def sort_mat_mx(df, min_size=15):
    '''
    df: Data-frame, should be sparse and binary and in the correct order.
    
    Function goes through and finds relatively non-overlapping rows and then
    outputs them in sorted order. 
    '''
    mat = df.copy()
    counts = mat.sum(1).astype(float)
    keep = []
    
    while len(mat) > 0:
        keep += [mat.index[0]]
        mat = (mat - mat.ix[0]).clip_lower(0)
        mat = mat[(mat.sum(1) > 5) * ((mat.sum(1) / counts).ix[mat.index] >= .5) * 
                  (counts.ix[mat.index] >= min_size)]
        
    mat = df.ix[keep].astype(int)
    o = np.sort(mat.apply(lambda s: ''.join(map(str, s)))).index[::-1]
    mat = mat[o]
    mat = mat.ix[:,mat.sum() != 0]
    return mat
```

```python
data_path = '/cellar/data/TCGA/Firehose__2012_10_24/'
cancers = os.listdir(data_path + 'ucsd_analyses')
def retreive_paths_for_datatype(data_type, report_ext, folder='Clinical', archive='ClinicalObject.p'):
    clinical = {}
    for cancer in cancers:
        clinical_obj = data_path + '/'.join(['ucsd_analyses', cancer, folder, data_type + report_ext,
                                             archive])
        if os.path.isfile(clinical_obj):
            clinical[cancer] = clinical_obj
    return clinical
```
