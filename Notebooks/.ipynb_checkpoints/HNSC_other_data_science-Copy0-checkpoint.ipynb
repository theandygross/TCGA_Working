```python
def get_patient_set(filters):
    f1 = [age >= 100] + list(filters)
    filter_df = pd.concat(f1, axis=1)
    clinical_filter = filter_df.dropna().sum(1) == 0
    keepers_o = true_index(clinical_filter)
    keepers_o = keepers_o.intersection(mut.features.columns).intersection(cn.features.columns)
    
    f2 = [age >= 85] + list(filters)
    filter_df = pd.concat(f2, axis=1)
    clinical_filter = filter_df.sum(1) == 0
    keepers = true_index(clinical_filter)
    keepers = keepers.intersection(mut.features.columns).intersection(cn.features.columns)
    return keepers, keepers_o
```

```python
maf = pd.read_csv('/cellar/users/agross/Data/HNSC_Science/maf.csv')
maf = maf.dropna(how='all', axis=[0,1])
maf['Tumor_Sample_Barcode'] = map(lambda s: s.replace('-Tumor',''), maf['Tumor_Sample_Barcode'])
maf = maf.set_index(['Hugo_Symbol','Tumor_Sample_Barcode'])
maf = maf[maf.Validation_Status != 'Wildtype']
non_silent = maf[maf.Variant_Classification != 'Synonymous']
non_silent['counter'] = 1
hit_matrix = non_silent.counter.groupby(level=[0,1]).sum().unstack()
hit_matrix.columns = map(lambda s: s.replace('-Tumor',''), hit_matrix.columns)
hit_matrix = hit_matrix.fillna(0)
```

```python
def is_disruptive(s):
    if 'fs' in s:
        return False
    if len(s) != 7:
        #print s
        return False
    if lo.Polarity[s[6]] == 'stop':
        return True
    aa = s[3:6]
    if int(aa) in range(163,196) + range(236, 252):
        if lo.Polarity[s[2]] == lo.Polarity[s[6]]:
            return True
    return False

aa = pd.read_csv('/cellar/users/agross/Data/GeneSets/amino_acids.csv')
lo = aa.set_index('Symbol').groupby(level=0).first()

p53 = maf.ix['TP53'].reset_index()
status = pd.concat([p53.Protein_Change.dropna().map(is_disruptive), p53.Tumor_Sample_Barcode], axis=1, keys=['status','barcode']).set_index('barcode')['status']
status = (status == True).groupby(level=0).sum().clip_upper(1.)
status = status.ix[hit_matrix.columns].fillna(-1).map({-1:'WT',0:'Non-Disruptive',1:'Disruptive'})
```
