####Load API functions

```python
from Data.Firehose import read_rnaSeq
from Figures.Helpers import prettify_ax
from Figures.Boxplots import violin_plot_series
```

####Set path to data

```python
data_path = '/cellar/data/TCGA/Firehose__2013_04_21/'
```

####Read in data

```python
tissues = ['HNSC','LUSC','LUAD','THCA']
data = {t: read_rnaSeq(data_path, t, tissue_code='All') 
        for t in tissues}
```

####Plot differential expression

```python
fig, axs = subplots(1,4, figsize=(12,4), sharey=True)
for i,(t,df) in enumerate(data.iteritems()):
    ax = axs[i]
    violin_plot_series(df.ix['ASPM'], ax=ax) #API call to plot
    prettify_ax(ax)  #API call to clean up axes

    ax.set_xlabel('')
    ax.set_xticklabels(['Tumor','Normal'])
    ax.set_xbound(-.5,1.5)
    ax.set_title(t)
    ax.set_ylabel('')
axs[0].set_ylabel('$Log_{2}$ ASPM mRNA');
```
