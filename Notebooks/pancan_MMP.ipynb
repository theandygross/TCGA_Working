```python
def initialize_cancer(run, cancer_type):
    cancer = Cancer(cancer_type, run)
    cancer.initialize_data(run, save=True)
    
    initialize_cn(cancer_type, run.report_path, 'CN_broad');
    initialize_mut(cancer_type, run.report_path, create_meta_features=True);
```

##Read in all of the data

```python
cancers = {}
clinical = {}
mut = {}
cn = {}
for cancer in run.cancers:
    if cancer == 'COAD':
        continue
    try:
        c = run.load_cancer(cancer)
        clinical[cancer] = c.load_clinical()
        mut[cancer] = c.load_data('Mutation')
        cn[cancer] = c.load_data('CN_broad')
        cancers[cancer] = c
    except:
        do_nothing = True
```

```python
rna = {}
mirna = {}
rppa = {}
for cancer in run.cancers:
    if cancer == 'COAD':
        continue
    try:
        c = cancers[cancer]
        rna[cancer] = c.load_data('mRNASeq')
        mirna[cancer] = c.load_data('miRNASeq')
        rppa[cancer] = c.load_data('RPPA')
    except:
        do_nothing = True
```

####Update to most recent clinical data

```python
from Data.ProcessClinical import format_survival_from_data_portal
for cancer,clin in clinical.iteritems():
    try:
        path  = '/cellar/users/agross/Data/TCGA/{}_Followup/'.format(cancer)
        survival, timeline = format_survival_from_data_portal(path, cancer)
        clin.survival = survival
        clin.artificially_censor(5)
        clinical[cancer] = clin
    except:
        do_nothing = True
```

```python
def ttr(s):
    a,b = [l[1] for l in list(s.groupby(level=1)) if l[0] in ['01','11']]
    a.index = a.index.droplevel(1)
    b.index = b.index.droplevel(1)
    return ttest_rel(a,b)
```

```python
def ppp(s):
    s = s.unstack()[['01', '11']].dropna()
    return spearman_pandas(s['01'], s['11'])
```
