```python
fmla = 'Surv(days, event) ~ feature'
surv = clinical.survival.event_free_survival_8y
cox = lambda s: get_cox_ph_ms(surv, s, return_val='model', formula=fmla)[3][0]
def log_rank(feature):
    if len(feature.unique()) == 1:
        return 0
    m = get_cox_ph_ms(surv, feature, return_val='model', formula=fmla)
    r_data = m.rx2('call')[2]
    s = survival.survdiff(robjects.Formula(fmla), r_data)
    chisq = s.rx2('chisq')[0]
    return chisq

def cox_screen(df):
    stat = df.apply(cox, 1)
    rr = pd.concat([pd.Series(chi2.sf(stat, 1), index=stat.index), stat], axis=1, keys=['p','chi2'])
    rr['q'] = bhCorrection(rr.p)
    rr = rr.sort('p')
    return rr

def lr_screen(df):
    stat = df.astype(float).apply(log_rank, 1)
    rr = pd.concat([pd.Series(chi2.sf(stat, 1), index=stat.index), stat], axis=1, keys=['p','chi2'])
    rr['q'] = bhCorrection(rr.p)
    rr = rr.sort('p')
    return rr
```

```python
exp_change = lambda s: anova(s.reset_index(level=1).ix[:,0], s.reset_index(level=1).ix[:,1])
var_change = lambda s: bartlett_pandas(s.reset_index(level=1).ix[:,0], s.reset_index(level=1).ix[:,1])
```

```python
df = FH.read_rnaSeq(cancer.name, run.data_path, tissue_code='All', average_on_genes=True)

has_data = ((df.T.xs('01', level=1).T < -1).sum(1)) < (df.T.groupby(level=1).size()['01'] - 10)
df = df.ix[has_data]
df = df[df.T.xs('01', level=1).std() > .5]

ch = df.apply(exp_change, 1)
singles = df.ix[true_index(bhCorrection(ch.p) < .01)]

ss = singles.apply(var_change, 1)
singles = singles.ix[true_index(bhCorrection(ss.p) < .01)]
singles = singles.T.xs('01', level=1).T
```

```python
rr = cox_screen(singles)
rr.head()
```

```python
def violin_plot_series(s, **kw_args):
    '''
    Wrapper for drawing a violin plot on a series with a multi-index.
    The second level of the index is used as the binning variable. 
    '''
    assert s.index.levshape[1] > 1
    violin_plot_pandas(pd.Series(s.index.get_level_values(1), s.index), s, **kw_args)
```
