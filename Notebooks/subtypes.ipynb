```python
run = get_run('/cellar/data/TCGA/Firehose__2013_04_21/', 'Run_all')

cancer = run.load_cancer('HNSC')
clinical = cancer.load_clinical()
global_vars = cancer.load_global_vars()

mut = cancer.load_data('Mutation')
mut.uncompress()
#meth = cancer.load_data('Methylation')
cn = cancer.load_data('CN_broad')
cn.uncompress()
rna = cancer.load_data('mRNASeq')
rppa  = cancer.load_data('RPPA')
mirna = cancer.load_data('miRNASeq')
age = clinical.clinical.age.astype(float)

hpv_all = pd.read_csv('/cellar/users/agross/hpv_summary_3_20_13_distribute.csv', index_col=0)
hpv = hpv_all.Molecular_HPV.map({0:'HPV-', 1:'HPV+'})
hpv.name = 'HPV'
```

```python
clinical.survival = survival
clinical.artificially_censor(5)
#age = timeline.age
```

```python
surv = clinical.survival.survival_5y
```

```python
tab = pd.read_csv('/cellar/users/agross/Data/HNSC_Science/TCGA_followup_UPMC.csv', parse_dates=[1,2])
tab.index = tab['TCGA #'].apply(lambda s: 'TCGA-CN-' + str(s))
days = (tab['Date of last followup'] - tab['Date of Surgery'])
tab['days'] = days.apply(lambda x: x / np.timedelta64(1, 'D'))
tab['event'] = 1.*(tab['status last followup'] == 'DECEASED')
```

```python
surv.ix[tab.index] = tab[['days','event']].stack()

years=5
df = surv.copy().unstack()
df['event'] = df.event * (df.days < int(365.25*years))
df['days'] = df.days.clip_upper(int((365.25*years)))
surv_5y = df.stack()
```

```python
path = '/cellar/users/agross/HNSC_new_maf.txt'
maf = pd.read_table(path, skiprows=2)
maf = maf.dropna(how='all', axis=[0,1])
maf = maf.set_index(['Hugo_Symbol','Tumor_Sample_Barcode'])
non_silent = maf[maf.Variant_Classification != 'Silent']
non_silent['counter'] = 1
hit_matrix = non_silent.counter.groupby(level=[0,1]).sum().unstack()
hit_matrix = FH.fix_barcode_columns(hit_matrix, tissue_code='01')
hit_matrix = hit_matrix.fillna(0)
```

```python
filters = pd.concat([age>=85], axis=1)
clinical_filter = filters.sum(1) == 0
clinical_filter = clinical_filter.ix[hpv.index]
keepers = true_index(clinical_filter)

p53_mut = 1.*(mut.df.ix['TP53', keepers].dropna() > 0)
p53_mut.name = 'p53_mut'
del_3p = cn.df.ix['Deletion'].ix['3p25.3', keepers].ix[0].dropna()
del_3p.name = 'del_3p'
#del_3p = cn.features.ix['Deletion'].ix[['3p25.3','3p14.2','3p12.2'], keepers]
#del_3p = -1*((del_3p<0).sum() > 0)

combo = combine((p53_mut + (hpv == 'HPV+'))>0, del_3p==-1)
combo = combine(p53_mut, del_3p==-1)
two_hit = 1.*(combo == 'both')
two_hit.name = 'two_hit'
p53_mut = p53_mut.ix[two_hit.index]
del_3p = del_3p.ix[two_hit.index]

subtype = combine(two_hit, mut.features.ix['REACTOME_SHC_MEDIATED_SIGNALLING']).dropna() == 'REACTOME_SHC_MEDIATED_SIGNALLING'
#subtype = combine(two_hit, mut.df.ix['HRAS']>0).dropna() == 'HRAS'
df = pd.concat([3.*(combine(subtype, two_hit==0)=='both'), 4.*(combine(mut.features.ix['MUC5B']==1, two_hit==1)=='both'), (two_hit + 1)], axis=1)
subtypes = df.max(1)
subtypes = subtypes.map({1:2, 2:3, 3:1,4:4})
subtypes.name = 'subtypes'
```
