###Previously Defined Subtypes

```python
p53_mut = 1.*(mut.df.ix['TP53', keepers].dropna() > 0)
p53_mut.name = 'p53_mut'
del_3p = cn.df.ix['Deletion'].ix['3p14.2', keepers].ix[0].dropna()
del_3p.name = 'del_3p'

combo = combine(p53_mut==1, del_3p==-1)
two_hit = 1.*(combo == 'both')
two_hit.name = 'two_hit'

subtype = combine(two_hit, mut.features.ix['REACTOME_SHC_MEDIATED_SIGNALLING']).dropna() == 'REACTOME_SHC_MEDIATED_SIGNALLING'
df = pd.concat([3.*(combine(subtype, two_hit==0)=='both'), 4.*(combine(mut.features.ix['MUC5B']==1, two_hit==1)=='both'), (two_hit + 1)], axis=1)
subtypes = df.max(1)
```

###Run miRNA Data

```python
exp_change = lambda s: anova(pd.Series(s.index.get_level_values(1), s.index), s)
var_change = lambda s: bartlett_pandas(pd.Series(s.index.get_level_values(1), s.index), s)
```

```python
df = mirna.df.ix[:,keepers]
df.T.xs('01', level=1).std().hist()
```

```python
df = mirna.df.ix[:,keepers]
df = df.xs('01', level=1, axis=1)
binary = df > -3
binary = binary[binary.sum(1).isin(range(20, len(two_hit)/2))]
```

```python
real = df.ix[df.index.diff(binary.index)]
singles = real
singles = singles[((singles.max(1) - singles.min(1)) > 2)]
singles = singles[(singles.std(1) > .5)]
ch = mirna.df.ix[singles.index].apply(exp_change, 1)
singles = singles.ix[true_index(ch.p < .01)]
```

```python
df = real.ix[df.index.diff(singles.index)].dropna()
df = ((df.T - df.mean(1)) / df.std(1)).T
U, S, pc = frame_svd(df)

background = pc[0]
pc.ix[:,:5].apply(cox, args=(surv,)).T
```

```python
ss = screen_feature(background, spearman_pandas, singles)
```

```python
singles = singles.ix[ss.p > 10e-5]
df = ((df.T - df.mean(1)) / df.std(1)).T
U, S, pc = frame_svd(df)

pc.ix[:,:5].apply(cox, args=(surv,)).T
```
