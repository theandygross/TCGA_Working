```python
fmla = 'Surv(days, event) ~ feature'
surv = clinical.survival.survival_5y
cox = lambda s: get_cox_ph_ms(surv, s, return_val='model', formula=fmla)[3][0]
def log_rank(feature):
    if len(feature.unique()) == 1:
        return 0
    m = get_cox_ph_ms(surv, feature, return_val='model', formula=fmla)
    r_data = m.rx2('call')[2]
    s = survival.survdiff(robjects.Formula(fmla), r_data)
    chisq = s.rx2('chisq')[0]
    return chisq

def cox_screen(df):
    stat = df.apply(cox, 1)
    rr = pd.concat([pd.Series(chi2.sf(stat, 1), index=stat.index), stat], axis=1, keys=['p','chi2'])
    rr['q'] = bhCorrection(rr.p)
    rr = rr.sort('p')
    return rr

def lr_screen(df):
    stat = df.astype(float).apply(log_rank, 1)
    rr = pd.concat([pd.Series(chi2.sf(stat, 1), index=stat.index), stat], axis=1, keys=['p','chi2'])
    rr['q'] = bhCorrection(rr.p)
    rr = rr.sort('p')
    return rr
```

```python
def extract_pc(df, pc_threshold=.2, standardize=True):
    if standardize:
        df = ((df.T - df.mean(1)) / df.std(1)).T
    try:
        U,S,vH = frame_svd(df)
    except np.linalg.LinAlgError:
        return None
    p = S**2/sum(S**2)
    pat_vec = vH[0]
    gene_vec = U[0]
    pct_var = p[0]
    if sum(gene_vec) < 0:
        gene_vec = -1*gene_vec
        pat_vec = -1*pat_vec
    ret = {'pat_vec': pat_vec, 'gene_vec': gene_vec, 'pct_var': pct_var}
    return  ret if pct_var > pc_threshold else None

def extract_pc_fancy(df, pc_threshold=.2):
    tt = df.xs('11', axis=1, level=1)
    m, s = tt.mean(1), tt.std(1)
    rr = df.apply(exp_change, 1).sort('p')
    ddd = df.xs('01', axis=1, level=1)
    ddd = ((ddd.T - m) / s).T
    ddd = ddd.ix[true_index(rr.p < .05)]
    pc = extract_pc(ddd, pc_threshold, standardize=False)
    return pc

def extract_geneset_pcs(df, gene_sets):
    '''Extract PCs for all gene sets.'''
    pc = {p: extract_pc_fancy(df.ix[g].dropna(), pc_threshold=0) for p,g, in gene_sets.iteritems()}
    pc = DataFrame({p: s for p,s in pc.iteritems() if s}).T
    pat_vec = DataFrame(pc.pat_vec.to_dict()).T
    return pc.gene_vec, pc.pct_var, pat_vec

#batch = batch.ix[:,'01']
batch.name = 'batch'
def batch_stratified_cox(f):
    batch.name = 'batch'
    fmla = 'Surv(days, event) ~ feature + strata(batch)'
    chi2 = get_cox_ph_ms(surv, f, covariates=pd.concat([batch], axis=1),  return_val='model', formula=fmla)[3][0]
    return chi2
def cox_screen(df):
    stat = df.apply(cox, 1)
    rr = pd.concat([pd.Series(chi2.sf(stat, 1), index=stat.index), stat], axis=1, keys=['p','chi2'])
    rr['q'] = bhCorrection(rr.p)
    rr = rr.sort('p')
    return rr
```
