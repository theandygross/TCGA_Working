```python
from scipy.stats import f_oneway
def anova(hit_vec, response_vec):
    hit_vec = hit_vec.reindex_like(response_vec)
    return f_oneway(*[response_vec[hit_vec == num].dropna() 
                      for num in set(hit_vec.dropna())])[1]
```

```python
df =  pandas.DataFrame(dict((m,dict((p, anova(mut_vec, exp_vec)) for p,exp_vec in exp_df.iterrows()))
                       for m, mut_vec in mut_df.iterrows()))
```

```python
from Processing.Helpers import match_series
def split_a_by_b(a,b):
    a, b = match_series(a, b)
    groups = [a[b==num] for num in set(b)]
    return groups
```

```python
from Reports.Figures import *
def histo_compare(hit_vec, response_vec, ax=None):
    '''
    Split response_vec by hit_vec and compared histograms.  
    Also plots the kde of the whole response_vec.
    '''
    if ax is None:
        fig, ax = plt.subplots(1,1, figsize=(5,3))
    else:
        fig = fig.gcf()
    kde1 = gaussian_kde(response_vec)
    x_eval = linspace(min(response_vec), max(response_vec), num=200)
    ax.plot(x_eval, kde1(x_eval), 'k-')
    miss, hit = split_a_by_b(response_vec, hit_vec)
    ax.hist(miss, bins=20, normed=True, alpha=.2, label='WT');
    ax.hist(hit, bins=10, normed=True, alpha=.5, label='Mut');
    ax.legend()
    return fig
```
