```python
import pickle as pickle
import os as os
from collections import defaultdict

from pandas import Series, DataFrame, MultiIndex

def tree(): return defaultdict(tree)
```

```python
data_path = '/cellar/data/TCGA/Firehose__2012_10_24/'
cancers = os.listdir(data_path + 'ucsd_analyses')
def retreive_paths_for_report_ext(report_ext, folder='Clinical', archive='ClinicalObject.p'):
    clinical = tree()
    for cancer in cancers:
        path = data_path + '/'.join(['ucsd_analyses', cancer, folder, ''])
        for dt in os.listdir(path):
            if dt.split('_')[-1] == report_ext:
                clinical_obj = path + '/'.join([dt, archive])
                if os.path.isfile(clinical_obj):
                    clinical[cancer][dt.split('_')[0]] = clinical_obj
    return DataFrame(clinical)
```

```python
def get_series_lists(cancer):
    def stack_me(df, event_type, statistic):
        idx = MultiIndex.from_tuples([(cancer.cancer, cancer.data_type, event_type, statistic, p) for p in df.index])
        idx.names = ['cancer', 'data_type', 'event_type', 'statistic', 'event_name']
        df = DataFrame(df.as_matrix(), index=idx, columns=df.columns)
        df.columns.name = 'test'
        s = df.stack()
        return s
    
    if hasattr(cancer, 'q_pathways') and (len(cancer.q_pathways) > 0):
        s_list = [stack_me(cancer.q_pathways, 'pathway', 'q'),
                  stack_me(cancer.p_pathways, 'pathway', 'p')]
    if hasattr(cancer, 'q_genes') and (len(cancer.q_genes) > 0):
        s_list += [stack_me(cancer.q_genes, 'gene', 'q'),
                   stack_me(cancer.p_genes, 'gene', 'p')]
    return s_list
```

```python
paths = retreive_paths_for_report_ext('LR')

stackify = lambda df: df.astype(float).stack().replace(0, nan).dropna()

def compress_cancer(cancer):
    if cancer.data_type in ['mutation', 'deletion', 'amplification']:
        cancer.hit_matrix = stackify(cancer.hit_matrix)
        cancer.meta_matrix = stackify(cancer.meta_matrix)
        cancer.single_matrix = stackify(cancer.single_matrix)
    else:
        keepers = set.union(*[cancer.gene_sets[p] for p in cancer.q_pathways.index])
        cancer.data_matrix = cancer.data_matrix.ix[keepers]
    return cancer
```

```python
def frame_svd(data_frame, impute='mean'):
    '''
    Wrapper for taking in a pandas DataFrame, preforming SVD
    and outputting the U, S, and vH matricies in DataFrame form.
    '''
    if impute == 'mean':
        data_frame = data_frame.dropna(thresh=int(data_frame.shape[1] * .75))
        data_frame = data_frame.dropna(how='all', axis=1)
        data_frame = data_frame.fillna(data_frame.mean())
    
    U,S,vH = svd(data_frame.as_matrix(), full_matrices=False)
    U = DataFrame(U, index=data_frame.index)
    vH = DataFrame(vH, columns=data_frame.columns).T
    return U,S,vH

def get_entropy(df):
    s = frame_svd(df)[1]
    p = s**2 / (sum(s**2))
    entropy = -sum(p*log(p)) / log(len(s))
    return entropy

```

```python
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as dist
def fancy_raster(df, cluster=False, cmap=cm.Spectral, norm=None):
    fig, ax = plt.subplots(1,1, figsize=(12,8))
    df = df.fillna(0).T
    if cluster:
        d = dist.pdist(df, 'cosine')
        D = dist.squareform(d)
        Y = sch.linkage(D)
        Z = sch.dendrogram(Y, no_plot=True)
        order = Z['leaves']
        
        df = df.ix[order]
        icoord = array(Z['icoord']) - array(Z['icoord']).min()
        icoord = icoord * ((len(Z['leaves']) - 1) / icoord.max())
    
        dcoord = array(Z['dcoord']) + df.shape[1] - .3
        for i,z,c in zip(icoord, dcoord, Z['color_list']):
            ax.plot(z,i,color=c, lw=2, alpha=.8)
            
        ax.tick_params(axis='x', top='off')
        if all(df.columns == df.index):
            df = df.ix[order, order]        
        else:
            df = df.T
            d = dist.pdist(df, 'cosine')
            D = dist.squareform(d)
            Y = sch.linkage(D)
            Z = sch.dendrogram(Y, no_plot=True)
            
            order = Z['leaves']
            df = df.ix[order].T
            icoord = array(Z['icoord']) - array(Z['icoord']).min()
            icoord = icoord * ((len(Z['leaves']) - 1) / icoord.max())
        
            dcoord = -1*array(Z['dcoord']) - .7 
            for i,z,c in zip(icoord, dcoord, Z['color_list']):
                ax.plot(i,z,color=c, lw=2, alpha=.8)
                
            ax.tick_params(axis='x', top='off')
            ax.tick_params(axis='y', right='off')
        
    
    if cluster:
        
        ax.set_frame_on(False)
    df[df==0] = nan
    img = ax.imshow(df, interpolation='Nearest', cmap=cmap, norm=norm)    
    
    ax.set_yticks(range(len(df.index)))
    ax.set_yticklabels(df.index)
    ax.set_xticks(arange(len(df.columns)))
    ax.set_xticklabels(df.columns, rotation=360-90, ha='center');
    ax.hlines(arange(len(df.index)-1)+.5, -.5, len(df.columns)-.5, 
              color='white', lw=6)
    ax.vlines(arange(len(df.columns)-1)+.5, -.5, len(df.index)-.5, 
              color='white', lw=6)
    return img
```
