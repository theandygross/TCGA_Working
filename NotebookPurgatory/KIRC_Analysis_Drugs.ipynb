```python
drugs = {'VEGF': ['sunitinib', 'sorafenib', 'avastin', 'pazopanib', 'sutent', 'tarceva','nexavaar',
         'sutent (sunitinib)', 'gefitinib', 'nexavar', 'bevacizumab', 'inf', 'bay-439006', 
         'azd', 'iressa', 'sorafenib - nexavar', 'axitinib', 'sunitinib (sutent)', 'tipifarnib',
         'tyrosine kinase inhibitor', 'votrient', 'zd6474'],
         'mTOR': ['temsirolimus','everolimus','proleukin','rad001','torisel','afinitor'],
         'IL2': ['interferon','gemcitabine','xeloda','oncophage','il-2','il-2 (high dose)','proleukin (il-2)',
          'oncophage vaccine','interleukin-2','interferon-alpha','interferon alpha','thalidomide',
          'intron a','nab-rapamycin','capecitabin','perifosine','capecitabine','gemzar','alpha interferon',
          'roferon-a','5-fluorouracil','il-2 thearpy (interleukin)','high dose interleukin-2',
          'ifn-alpha (intron)', 'gemictiabine', 'bortezomib','interleukin 2-high dose']}
drug_map = {drug:family for family, drugs in drugs.items() for drug in drugs}
```

```python
result_path = '/scratch/TCGA/Firehose__2012_01_16/ucsd_analyses'
run = sorted(os.listdir(result_path))[1]
run = pickle.load(open('/'.join([result_path, run, 'RunObject.p']), 'rb'))
```

```python
survival_test = 'event_free_survival_5y'
covariates = ['age', ('mutation', 'rate_non')]
cov_df = global_vars.join(clinical.clinical, how='outer').join(cdk_del)
cov_df = cov_df[covariates]
remerge = lambda s: '__'.join(s) if type(s) != str else s
cov_df = cov_df.rename(columns=remerge)
surv = clinical.survival[survival_test]
test = SurvivalTest(surv, cov_df)
test.name = survival_test
test.check_feature = lambda s: True
```

```python
def draw_me(f):
    draw_survival_curves(mut.features.ix[f, df.columns], surv, ann='p')
    figsize=(6,4)
    pathway_plot(mut.df.ix[run.gene_sets[f], df.columns], plt.gca())
    plt.tight_layout()
    plt.savefig('tmp1.png', dpi=75, bbox_inches=0, pad_inches=0)
    plt.close('all')
    return side_by_side(['tmp.png', 'tmp1.png'])
```

```python
df = rppa.features.ix[:,drug_given[drug_given.IL2].index]
df = df.dropna(axis=1)
counts = Series((df != 0).sum(1), name='counts')
df  = df[counts > 5]
```

```python
survival_test = 'event_free_survival_5y'
covariates = ['age']
cov_df = global_vars.join(clinical.clinical, how='outer').join(cdk_del)
cov_df = cov_df[covariates]
remerge = lambda s: '__'.join(s) if type(s) != str else s
cov_df = cov_df.rename(columns=remerge)
surv = clinical.survival[survival_test]
test = SurvivalTest(surv, cov_df)
test.name = survival_test
test.check_feature = lambda s: True
```

```python
il2_rna = run_feature_matrix(df, test)
```

```python
il2s = il2_rna.join(counts).sort(columns=[('Univariate','p')])
il2s.head(10)
```
