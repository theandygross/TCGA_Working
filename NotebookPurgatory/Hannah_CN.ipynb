```python
def map_chr(data, chr_num):
    chr_map = data[data.chrom == chr_num]
    chr_map = chr_map.set_index('chr_start')
    m = chr_map.adjusted_log_ratio.groupby(lambda s: s/10000)
    m = m.agg({'size': len, 'median': pd.Series.median})
    m = m[m.size > 20]['median']
    m.index = m.index * 10000
    return m
```

```python
f = '/cellar/users/agross/Downloads/possible_SVs.csv'
sv = pd.read_csv(f)

sv_comp = sv.ix[:,6:]
sv = sv.ix[:,:6]
sv.columns = pd.MultiIndex.from_tuples([('s1','chr'),('s1','pos'),('s1','orientation'), 
                           ('s2','chr'),('s2','pos'),('s2','orientation')])
sv = sv.stack(level=0)
sv.chr = sv.chr.map(lambda s: s[3:])
```

```python
def process(g):
    d = chr_pos[chr_pos.chr == g.ix['chr']]
    d =  (d[['start','end']] - g.ix['pos'])
    d = d[(d.start < 10000) * (d.end > -10000)]
    try:
        closest = ((d.start * -1)  + d.end).idxmax()
        d1 = d.ix[closest]
        d1 = d1.set_value('gene', closest)  
    except:
        d1 = pd.Series({'gene': nan, 'start': nan, 'end': nan})
    g = g.append(d1)
        
    return g
```

```python
sv = sv.apply(process, axis=1)
sv.index = pd.MultiIndex.from_tuples(sv.index)
sv = sv[['gene','chr','pos','start','end','orientation']]
```

```python
s = sv.dropna().groupby(level=0).size()
gv = sv[sv.index.get_level_values(0).isin(s[s > 1].index)]
```

```python
def sort_by_pos(g):
    g = g.unstack()
    g = g.sort(columns=['chr','pos'])
    g.index = ['s1','s2']
    return g.T.stack(dropna=False)
```

```python
sv = sv.reset_index().sort(columns=['level_0', 'chr','pos']).set_index(['level_0', 'level_1'])
sv.chr = sv.chr.map(lambda s: '0'+s if len(s) ==1 else s)
```

```python
ss = sv.unstack(level=1).join(sv_comp).sort(columns=[('chr','s1')])
ss.to_csv('svs_mapped_to_genes.csv')
```
