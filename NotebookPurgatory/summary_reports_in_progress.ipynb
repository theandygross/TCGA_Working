```python
def get_gene_lookup(gene_sets):
    gene_lookup = defaultdict(set)
    for pathway, gene_set in gene_sets.iteritems():
        for gene in gene_set:
            gene_lookup[gene].add(pathway)
    return gene_lookup
```

```python
from Reports.Reports import *
from Reports.Figures import *
from Processing.Tests import *

def sort_mat_mx(df, min_size=15):
    '''
    df: Data-frame, should be sparse and binary and in the correct order.
    
    Function goes through and finds relatively non-overlapping rows and then
    outputs them in sorted order. 
    '''
    mat = df.copy()
    counts = mat.sum(1).astype(float)
    keep = []
    
    while len(mat) > 0:
        keep += [mat.index[0]]
        mat = (mat - mat.ix[0]).clip_lower(0)
        mat = mat[(mat.sum(1) > 5) * ((mat.sum(1) / counts).ix[mat.index] >= .5) * 
                  (counts.ix[mat.index] >= min_size)]
        
    mat = df.ix[keep].astype(int)
    o = np.sort(mat.apply(lambda s: ''.join(map(str, s)))).index[::-1]
    mat = mat[o]
    mat = mat.ix[:,mat.sum() != 0]
    return mat
```

```python
def recompute_survival_models(cancer, test):
    pathways = cancer.p_pathways[test].order().dropna()
    pathways = pathways[cancer.q_pathways[test] < .2]
    
    get_surv = lambda vec: get_cox_ph_ms(cancer.clinical, vec, return_val='LR',
                                         **cancer.survival_tests[test])
    df = DataFrame({p: get_surv(cancer.meta_matrix.ix[p]) for p in pathways.index}).T
    df = df[df.LR < .05]
    models = df['model']
    df = df[['LR','hazzard','pathway_p']]
    annotation = DataFrame(dict((p,get_pathway_annotation_vec(p, cancer)) 
                               for p in df.index)).T
    df = annotation.join(df)
    return df, models
```

```python
class NozzleTable2(object):
    def __init__(self, table, caption='', idx_name=''):
        table = table.applymap(lambda s: "%8.2e"%s).astype(float)
        index_series = Series(dict((i,i) for i in table.index), name=idx_name)
        self.table = table.join(index_series)
        self.table = roll_df(self.table,1)
        self.row_pos = dict((g,i+1) for i,g in enumerate(self.table.index))
        self.col_pos = dict((c,i+1) for i,c in enumerate(self.table.columns))
        
        r_table = self.table.copy().fillna(1.)
        r_table = com.convert_to_r_dataframe(r_table) #@UndefinedVariable
        self.nozzle_table = nz.newTable(r_table, caption)
        
    def add_to_table(self, result, row, col, captions='', title='',
                     isSignificant=True):
        '''
        Add a figure from a file to the Nozzle table.
        This is just a wrapper to put a figure(s) in a result section.
        '''
        self.nozzle_table = nz.addTo(self.nozzle_table, result, 
                                     row=self.row_pos[row], 
                                     column=self.col_pos[col]) 
```

```python
def make_report(cancer):    
    if hasattr(cancer, 'p_genes'):
        if hasattr(cancer, 'filter_bad_pathways'):
            gene_lookup = get_gene_lookup(cancer.gene_sets)
            cancer.filter_bad_pathways(gene_lookup)
            cancer.q_pathways = cancer.q_pathways.fillna(1.)
            
    fig_path = cancer.data_path + cancer.report_folder + FIG_EXT 
    val = ('event_free_survival_5y' if 'event_free_survival_5y' in cancer.clinical 
           else 'survival_5y')
    
    df, models = recompute_survival_models(cancer, val)   
    
    bad_guys = df[df.hazzard > 1].sort(columns='LR')
    mat = sort_mat_mx(cancer.meta_matrix.ix[bad_guys.index], min_size=0)
    bad_guys = df.ix[mat.index]
    
    days_normed = (cancer.clinical.days_5y[mat.columns] / (365.*5))
    f = draw_pathway_overlaps(mat, days_normed, fig_path + 'bad_guys.png');
    ax = f.get_axes()[0]
    ax.set_yticks([])
    ax.set_xbound(lower=-.4)
    f.savefig(fig_path + 'bad_guys_nl.png')
    
    good_guys = df[df.hazzard < 1].sort(columns='LR')
    mat = sort_mat_mx(cancer.meta_matrix.ix[good_guys.index], min_size=0)
    days_normed = (cancer.clinical.days_5y[mat.columns] / (365.*5))
    good_guys = df.ix[mat.index]
    
    f = draw_pathway_overlaps(mat, days_normed, fig_path + 'good_guys.png');
    ax = f.get_axes()[0]
    ax.set_yticks([])
    ax.set_xbound(lower=-.4)
    f.savefig(fig_path + 'good_guys_nl.png')
    
    def add_cox_result(tab, p):
        f = fig_path + p + '_' + val + '.png'
        create_figure_bool(cancer, val, cancer.meta_matrix.ix[p], f)
        curves = nz.newFigure(f)
        m = base.summary(models[p])
        coef = com.convert_robj(m.rx2('coefficients'))
        coef = NozzleTable2(coef, caption='Model Coefficients', idx_name='covariate')
        f = fig_path + p + '.png'
        create_figure_bool(cancer, 'pathway_bar', cancer.meta_matrix.ix[p], f)
        counts = nz.newFigure(f)
        result = nz.newResult('')
        result = nz.addTo(result, nz.addTo(nz.newSubSection('Cox Model for ' + p), curves,
                            coef.nozzle_table), counts)
        tab.add_to_table(result, p, 'Pathway') 
        
    #fig_path_rel = '/'.join(['..'] + cancer.report_folder.split('/')[-2:]) + FIG_EXT    
    fig = nz.newFigure(fig_path + 'good_guys_nl.png', 
                       fileHighRes=fig_path + 'good_guys.png')
    tab = NozzleTable2(good_guys, idx_name='Pathway')
    
    for p in tab.table.index:
        add_cox_result(tab, p)
    
    good_section = nz.newSection('Good Guys')
    good_section = nz.addTo(good_section, fig, tab.nozzle_table)
    #report = nz.addToResults(report, good_section)
    
    fig = nz.newFigure(fig_path + 'bad_guys_nl.png', 
                       fileHighRes=fig_path + 'bad_guys.png')
    tab = NozzleTable2(bad_guys, idx_name='Pathway')
    
    for p in tab.table.index:
        add_cox_result(tab, p)
    
    bad_section = nz.newSection('Bad Guys')
    bad_section = nz.addTo(bad_section, fig, tab.nozzle_table)
    #report = nz.addToResults(report, bad_section)
    return good_section, bad_section
    #nz.writeReport(report, filename=cancer.data_path + cancer.report_folder + 'survival')
```

```python
def generic_header(report, prev_cancer, next_cancer, 
                   address='index.html'):
    report = nz.setMaintainerName(report, 'Andrew Gross')
    report = nz.setMaintainerEmail(report, "agross@ucsd.edu" );
    report = nz.setMaintainerAffiliation(report, 'UCSD- Bioinf. and '
                                                + 'Systems Biology' ); 
    next_page = '/'.join(['..', next_cancer, address])  
    prev_page = '/'.join(['..', prev_cancer, address])  
    report = nz.setPreviousReport(report, next_page, next_cancer)
    report = nz.setNextReport(report, prev_page, prev_cancer)
    return report
```

```python
cancers = array(report_sections.keys())
lookup = {c: cancers[array([i-1, i+1]) % len(cancers)] for i,c in 
          enumerate(cancers)}
```

```python
for cancer, cancer_report in report_sections.iteritems():
    report = nz.newReport('Report for ' + cancer)
    prev_cancer, next_cancer = lookup[cancer]
    report = generic_header(report, next_cancer, prev_cancer, 
                            address='survival_overlap_plots.html')  
    
    for data_type, reports in cancer_report.iteritems():
        stuff = nz.newSection(data_type)
        stuff = nz.addTo(stuff, reports['good_guys'], reports['bad_guys'])
        report = nz.addToResults(report, stuff)
    
    nz.writeReport(report, filename= data_path + '/'.join(['ucsd_analyses', cancer, 'survival_overlap_plots']))
```
