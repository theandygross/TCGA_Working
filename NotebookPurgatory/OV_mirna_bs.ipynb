```python
import Data.Firehose as FH
from scipy.stats import chi2

fmla = 'Surv(days, event) ~ feature'
surv = clinical.survival.survival_7y
cox = lambda s: get_cox_ph_ms(surv, s, return_val='model', formula=fmla)[3][0]
```

```python
df = FH.read_miRNASeq(cancer.name, run.data_path)
real = df[((df.max(1) - df.clip_lower(0).min(1)) > .5)]
real  = real[real.std(1) > .1]
real = df
```

```python
lr = real.apply(cox, axis=1)
```

```python
rr = pd.concat([pd.Series(chi2.sf(lr, 1), index=lr.index), lr], axis=1, keys=['p','LR'])
rr['q'] = bhCorrection(rr.p)
```
