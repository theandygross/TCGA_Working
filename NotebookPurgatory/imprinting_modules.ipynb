```python
from NetworkModule.ModuleFinding import *
from NetworkModule.Figures import *

def cluster_on_graph(data_vec, graph_influence, genes, average_degree=4.):
    ll = np.log(data_vec).ix[genes]
    influence = graph_influence.mul(ll, level=0).mul(ll, level=1)
    del graph_influence
    
    sparsity  = 2*average_degree/influence.index.levshape[0]
    cutoff = influence.quantile(1 - sparsity)
    influence = influence[influence > cutoff]
    
    Gr = nx.Graph()
    Gr.add_weighted_edges_from(((i,j,k) for (i,j),k in influence.iteritems()))
    
    membership = pre_cluster(Gr) 
    membership = membership[membership > 0]
    
    candidates = get_candidates(membership, Gr)
    candidates = map(tuple, candidates)
    return candidates, influence
```
