```python
mut = Dataset(cancer, r, 'MAF')
hit_matrix = mut.df.fillna(0).clip_upper(1.)
meta_matrix = DataFrame({p: mut.df.ix[g].sum() for p,g in gene_sets.iteritems()}).T
meta_matrix = meta_matrix.fillna(0).clip_upper(1.)
meta_matrix = meta_matrix[meta_matrix.sum(1) > 15]
```

```python
def is_one_gene(p):
    counts = hit_matrix.ix[gene_sets[p]].sum(1).dropna().order()
    with_top = hit_matrix.ix[gene_sets[p]].sum().clip_upper(1).sum()
    without = hit_matrix.ix[gene_sets[p] - {counts.idxmax()}].sum().clip_upper(1).sum()
    return ((with_top - without) / without) > .5
```

```python
s = Series({p: is_one_gene(p) for p in meta_matrix.index})
meta_matrix = meta_matrix.ix[s==False]
```

```python
counts = (meta_matrix.sum(1)  - (meta_matrix.shape[1] / 2)) / meta_matrix.shape[1]
counts = counts.abs().order()

trash = []
for p, vec in meta_matrix.iterrows():
    if vec.value_counts().index[0] == 1:
        vec = (-1*vec) + 1
    mat = (3*vec - meta_matrix)
    frac = ((mat == 2).sum(1).astype(float) / (vec == 1).sum()).order()
    if frac[-2] > .5:
        if counts[p] < counts[frac.index[-2]]:
            trash += [p]
```

```python
meta_matrix = meta_matrix.ix[list(set(meta_matrix.index) - set(trash))]
```

```python
from rpy2.robjects import r, FloatVector
def bhCorrection(s, n=None):
    if n==None:
        n = len(s)
    return Series(r('p.adjust')(FloatVector(s), method='BH', n=n),index=s.index, 
                  name=s.name)
```

```python
from Reports.Reports import *
from Reports.Figures import *
from Processing.Tests import *

def sort_mat_mx(df, min_size=15):
    '''
    df: Data-frame, should be sparse and binary and in the correct order.
    
    Function goes through and finds relatively non-overlapping rows and then
    outputs them in sorted order. 
    '''
    mat = df.copy()
    counts = mat.sum(1).astype(float)
    keep = []
    
    while len(mat) > 0:
        keep += [mat.index[0]]
        mat = (mat - mat.ix[0]).clip_lower(0)
        mat = mat[(mat.sum(1) > 5) * ((mat.sum(1) / counts).ix[mat.index] >= .5) * 
                  (counts.ix[mat.index] >= min_size)]
        
    mat = df.ix[keep].astype(int)
    o = np.sort(mat.apply(lambda s: ''.join(map(str, s)))).index[::-1]
    mat = mat[o]
    mat = mat.ix[:,mat.sum() != 0]
    return mat
```
