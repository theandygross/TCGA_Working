```python
import pickle as pickle
import subprocess
import os
from pandas import Panel

def run_perm(mat, test, N=10, max_processes=7):    
    patients = list(mat.columns)
    perm_test = DataFrame({i: random.permutation(patients) for i in range(N)}, index=patients)
    del test.first_pass
    del test.full_test
    pickle.dump(test, open('test.p','wb'))
    
    def run_perm(mat, label):
        perm_file = 'mat_' + str(label) + '.p'
        pickle.dump(mat, open(perm_file, 'wb'))
        r = subprocess.Popen(['nice', '-n', '5', 'python', './testing.py', 
                             perm_file])
        return r
    
    processes = set()
    for i,perm in perm_test.iteritems():
        processes.add(run_perm(mat.rename(columns=perm), i))
        if len(processes) >= max_processes:
            os.wait()
            processes.difference_update(
                set([p for p in processes if p.poll() is not None]))
            
    if len(processes) > 0:
        os.wait()     
        
    m= {}
    for i,perm in perm_test.iteritems():
        m[i] = pickle.load(open('res_mat_' + str(i) + '.p','rb'))
    m = Panel(m)
    return m
```
