####plot_guy just takes an coordinate and an option for a frown
####plot_args are a dictionary of arguments to pass to line function

```python
import matplotlib.backend_bases as B
```

```python
def plot_guy(x, y, frown=False, **plot_args):
    an = array(linspace(0,2*pi,100))
    head = cos(an)+x, sin(an)+y + 5
    body = array([0,0])+x, array([4,2])+y
    legs = array([-1,0,1])+x, array([0,2,0])+y
    arms = array([-1.5,0,1.5])+x, array([3.5,3,3.5])+y
    eye1 = .2*cos(an)-.35 + x, .2*sin(an) + 5.2 + y
    eye2 = .2*cos(an)+.35 + x, .2*sin(an) + 5.2 + y
    plot(*(head + body + legs + arms + eye1 + eye2),
                  **plot_args)
    if frown:
        mouth, = plot(cos(an[15:35]) + x, sin(an[15:35]) + 3.8 + y, **plot_args)
    else:
        mouth, = plot(cos(an[65:85]) + x, sin(an[65:85]) + 5.4 + y, **plot_args)
colors = plt.rcParams['axes.color_cycle'][:4]
```
